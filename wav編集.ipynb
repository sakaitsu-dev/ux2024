```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io.wavfile import read, write 
```

```python
plt.rcParams['figure.figsize'] = (10,3)
wave_ax = {'xlabel':'Time[s]','ylabel':'Apmlitude[Pa]'}

fs = 48000
L = 0
```

# check
健全2枚,剥離2枚,各2回,計4回打診

```python
_, data = read('/workspace/original/check.wav')
data = data[:,L]
time = np.arange(len(data))/fs

_,ax = plt.subplots()
ax.plot(time,data)
ax.set(**wave_ax)
```

```python
cut_data = data[int(4*fs):int(7.5*fs)]
cut_time = np.arange(len(cut_data))/fs

_,ax = plt.subplots()
ax.plot(cut_time,cut_data)
ax.set(**wave_ax)
```

```python
write('workspace/sound/check.wav',fs,cut_data)
```
